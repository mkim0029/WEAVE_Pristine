# Stellar Spectrum Analysis and Plotting

This notebook demonstrates how to use the `spectrum_reader` module to load stellar spectrum data and create wavelength vs flux plots.

## 1. Import Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import h5py
from hdf5_spectrum_reader import HDF5SpectrumReader

# Set matplotlib style for better plots
plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
```

## 2. Load Spectrum Data .h5

```python
spectrum_file = "weave_nlte_grids.h5"
reader = HDF5SpectrumReader(spectrum_file)
```

## 3. Extract Wavelength and Flux Data

Extract wavelength and flux arrays from the loaded spectrum data and plot.

```python
spectrum_index = 99 #look at first spectrum
```

```python
# Extract wavelength and flux data
wavelength, flux = reader.get_wavelength_flux(spectrum_index) 
```

## 4. Create Basic Wave vs Flux Plot

Create a basic line plot showing wavelength on the x-axis and flux on the y-axis using matplotlib.

```python
plt.plot(wavelength, flux)
plt.xlabel('Wavelength ()', fontsize=12)
plt.ylabel('Flux', fontsize=12)

# Create title with stellar parameters
info = reader.get_spectrum_info(spectrum_index)
params = info['stellar_parameters']
title_parts = [f"Spectrum: {info['filename']}"]
        
# Add key stellar parameters to title
if params.get('teff') is not None:
    title_parts.append(f"T={params['teff']:.0f}K")
if params.get('log_g') is not None:
    title_parts.append(f"log g={params['log_g']:.2f}")
if params.get('metallicity') is not None:
    title_parts.append(f"[Fe/H]={params['metallicity']:.2f}")
        
title = " | ".join(title_parts)

plt.title(title, fontsize=14, pad=15)
plt.show()
```

Z-score (mean/std) normalisation

```python
norm_flux = (flux - flux.mean()) / flux.std()
plt.plot(wavelength, norm_flux)
plt.xlabel('Wavelength ()', fontsize=12)
plt.ylabel('Normalised Flux', fontsize=12)

# Create title with stellar parameters
info = reader.get_spectrum_info(spectrum_index)
params = info['stellar_parameters']
title_parts = [f"Spectrum: {info['filename']}"]
        
# Add key stellar parameters to title
if params.get('teff') is not None:
    title_parts.append(f"T={params['teff']:.0f}K")
if params.get('log_g') is not None:
    title_parts.append(f"log g={params['log_g']:.2f}")
if params.get('metallicity') is not None:
    title_parts.append(f"[Fe/H]={params['metallicity']:.2f}")
        
title = " | ".join(title_parts)

plt.title(title, fontsize=14, pad=15)
plt.show()
```
